```python
pred = np.array([5,7,3,5,4,1,3,7])
y = np.array([5,7,3,5,4,1,3,7])
```

```python
mse = (1/pred.size) * np.sum(np.square( pred-y ))
mse
```

```python
pred = np.array([5,7,3,5,4,4,3,7])
y = np.array([5,3,3,5,4,9,7,7])
```

```python
mse = (1/pred.size) * np.sum(np.square( pred-y ))
mse
```

```python
pred = np.array([5,7,3,5,4,4,3,7])
y = np.array([5,3,3,5,4,9,7,7])
```

```python
rmse = np.sqrt((1/pred.size) * np.sum(np.square( pred-y )))
rmse
```
