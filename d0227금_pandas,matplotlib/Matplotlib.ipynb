```python
def data_s(data) : 
    fig, ax = plt.subplots()
    
    ax.plot(data)
    ax.set_title(' ', fontdict={'family':'HYGungSo-Bold', 'color': 'red','size':30})
    ax.set_xlabel('', color = 'y')
    ax.set_ylabel('', color= 'g')
    return fig, ax

data_s(data)
```

```python
fig, ax = data_s(data)

ax.grid(True)
ax.plot(data, label='')
ax.legend(loc='upper center')

plt.show()
```

```python
data_s(data)
```
